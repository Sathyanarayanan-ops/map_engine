```python
import osmnx as ox 

# Specify the location (Blacksburg, Virginia)
place_name = "Blacksburg, Virginia, USA"

# Extract road network
road_network = ox.graph_from_place(place_name, network_type="drive")

# Save the road network as a GraphML file
ox.save_graphml(road_network, filepath="blacksburg_road_network.graphml")

print("Road network data for Blacksburg has been saved.")
```

```python
ox.plot_graph(road_network)
```
